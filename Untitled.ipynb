## Modelling

```python
# Creating the train set and test set
X = df_copy.drop('y' , axis = 1)
y = df_copy['y']

# Spliting the datasets for testing and training for evaluation 
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
# Models to test on
# Baseline scores

models = {'Logistic Regression' : LogisticRegression() , 
         'KNN' : KNeighborsClassifier() ,
         'SVM (Kernel:Linear)':SVC(kernel = 'linear') ,
          'SVM (Kernel:rbf)':SVC(kernel = 'rbf'),
         'Desicion Tree' : DecisionTreeClassifier()}
```

```python
def prep_models(models , X_train , X_test , y_train  , y_test):
    """
        fitting and testing different models 
        models = dict of the required models
    """
    
    # set the random seed
    np.random.seed(45)
    
    #scores dict 
    mod_scores = {}
    
    for name , model in models.items():
        model.fit(X_train , y_train) # fitting the model
        pred = model.predict(X_test)
        # evaluating and adding the scores to the dict
        mod_scores[name] = {'Precision' : precision_score(pred , y_test ) , 
                           'Recall' : recall_score(pred , y_test ) , 
                           'F1 Score' : f1_score(pred , y_test )  , 
                           'Accuracy' : accuracy_score(pred , y_test )}
    return mod_scores
```

```python
%%time
model_scores = prep_models(models , 
                           X_train = X_train ,
                           X_test = X_test , 
                          y_train = y_train , 
                          y_test = y_test)
model_display = pd.DataFrame(model_scores)
model_display
```

```python
model_display.T.plot(kind = 'bar',figsize=(15,6))
plt.title('Base-Line Models')
plt.legend(loc=(1.04 , 0.2))
plt.xlabel("MODELS")
plt.xticks(rotation = 0);
```

#### From the base line model it is clear that KNN and SVM(rbf)  are giving the best results
* These models are better as these models are having better accuracy along with better precision than the others 
* Although Logistic Regression and SVM linear kernel has 1 recall the accuracy and precision are not that good 

### Tuning the KNN and SVM (kernel rbf models ) as these models have the best results

```python
# tuning the KNN model to get better results

test_score = []
train_score = []

neighbors = range(1,21)

knn = KNeighborsClassifier()

for i in neighbors:
    np.random.seed(45)
    knn.set_params(n_neighbors = i)
    knn.fit(X_train , y_train)
    test_score.append(knn.score(X_test,y_test))
    train_score.append(knn.score(X_train , y_train))


plt.figure(figsize=(15,4))
plt.plot(neighbors , test_score , label = 'Test Scores')
plt.plot(neighbors , train_score , label = 'Train Scores')
plt.title('KNN hyperparameter tuning')
plt.xticks(np.arange(1,21,1))
plt.legend();

print(f'The best score is {max(test_score)*100:.2f}%')
```

```python
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
```

```python
np.random.seed(42)
rs_log = RandomizedSearchCV(SVC() , 
                           param_distributions=param_grid , 
                           cv = 5 , # cross validation 
                           n_iter= 20 , 
                           verbose=True,
                           refit = True)
rs_log.fit(X_train,y_train)
```

```python
rs_log.best_params_
```

```python
rs_log.score(X_test,y_test)
```

```python
print(confusion_matrix(y_test,pred))
def plot_conf_mat(x , y):
    """
    Plotting the confusion matrix using seaborn heatmap 
    """
    
    fig , ax = plt.subplots(figsize=(3,3))
    ax = sns.heatmap(confusion_matrix(x , y) , annot= True , cbar= False,fmt='g')
    plt.xlabel('True values')
    plt.ylabel('Predicted labels');
    
plot_conf_mat(y_test , pred)
```
